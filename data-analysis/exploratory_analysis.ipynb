```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
DATA_PATH = "../docs/master_property_tax_enforcement_dataset.csv"

master_property = pd.read_csv(DATA_PATH, low_memory=False)

print(master_property.shape)
display(master_property.head())
```

#### Section 1: Property Composition

What type of properties dominate the lien sale pool?

```python
# TAX CLASS

tax_class_summary = (
    master_property["tax_class_code"]
    .value_counts(dropna=False)
    .rename_axis("tax_class")
    .reset_index(name="properties")
)

tax_class_summary["share"] = (
    tax_class_summary["properties"] /
    tax_class_summary["properties"].sum()
)

display(tax_class_summary)
```

<B>
The final-sale pool is dominated by residential property. </B>

 Final-sale pool is:

48.6% Class 1 <BR>
23.6% Class 2 <br>
27.8% Class 4 <br>


```python

# Borough composition

borough_summary = (
    master_property["borough_lien"]
    .value_counts(dropna=False)
    .rename_axis("borough")
    .reset_index(name="properties")
)

borough_summary["share"] = (
    borough_summary["properties"] /
    borough_summary["properties"].sum()
)

display(borough_summary)
```

<b>
Brooklyn and Queens Dominate the Final-Sale Pool
</B> <BR>

Brooklyn: 38.1% <BR>
Queens: 29.1% <BR>
Bronx: 13.6% <BR>
Manhattan: only 10.9% <BR>

```python
# Building class composition

building_class_summary = (
    master_property["building_class"]
    .value_counts(dropna=False)
    .head(25)
    .rename_axis("building_class")
    .reset_index(name="properties")
)

building_class_summary["share"] = (
    building_class_summary["properties"] /
    building_class_summary["properties"].sum()
)

display(building_class_summary)
```

```python
# PLUTO land use composition

landuse_summary = (
    master_property["landuse"]
    .value_counts(dropna=False)
    .rename_axis("landuse")
    .reset_index(name="properties")
)

landuse_summary["share"] = (
    landuse_summary["properties"] /
    landuse_summary["properties"].sum()
)

display(landuse_summary)
```

```python
# Residential unit distribution

master_property["unitsres"] = pd.to_numeric(
    master_property["unitsres"],
    errors="coerce"
)

master_property["unitstotal"] = pd.to_numeric(
    master_property["unitstotal"],
    errors="coerce"
)

display(
    master_property[[
        "unitsres",
        "unitstotal"
    ]].describe()
)
```

```python
# Small homes vs larger buildings

master_property["small_property_flag"] = np.where(
    master_property["unitsres"].fillna(0) <= 3,
    1,
    0
)

small_property_summary = (
    master_property["small_property_flag"]
    .value_counts(dropna=False)
    .rename_axis("small_property_flag")
    .reset_index(name="properties")
)

small_property_summary["share"] = (
    small_property_summary["properties"] /
    small_property_summary["properties"].sum()
)

display(small_property_summary)
```

<b> Small Residential Buildings Dominate </b> <br>

C0/C1/C2/C3/C7 → walk-up apartments <br>
B1/B2/B3 → 2-family homes <br>
A1/A2/A5 → 1-family homes <br>
R4 → condo-related <br>
S2 → mixed-use residential  <br>

Most final-sale properties are very small properties (84%)

```python
# Property value distribution

value_cols = [
    "fullval",
    "avtot",
    "avland",
    "extot"
]

for col in value_cols:
    master_property[col] = pd.to_numeric(
        master_property[col],
        errors="coerce"
    )

display(
    master_property[value_cols].describe()
)
```

```python
# Top building classes by HPD distress

building_hpd_summary = (
    master_property
    .groupby("building_class", as_index=False)
    .agg(
        properties=("bbl", "nunique"),
        avg_open_hpd_violations=("open_hpd_violations", "mean"),
        avg_open_c_violations=("open_c_violations", "mean")
    )
    .sort_values(
        "avg_open_hpd_violations",
        ascending=False
    )
)

display(building_hpd_summary.head(20))
```

Older walk-up classes,older multifamily stock,
certain mixed-use classes,
and elevator/walk-up apartments

associated with extremely high violation counts.
