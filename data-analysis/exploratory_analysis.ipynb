```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
DATA_PATH = "../docs/master_property_tax_enforcement_dataset.csv"

master_property = pd.read_csv(DATA_PATH, low_memory=False)

print(master_property.shape)
display(master_property.head())
```

#### Section 1: Property Composition

What type of properties dominate the lien sale pool?

```python
# TAX CLASS

tax_class_summary = (
    master_property["tax_class_code"]
    .value_counts(dropna=False)
    .rename_axis("tax_class")
    .reset_index(name="properties")
)

tax_class_summary["share"] = (
    tax_class_summary["properties"] /
    tax_class_summary["properties"].sum()
)

display(tax_class_summary)
```

<B>
The final-sale pool is dominated by residential property. </B>

 Final-sale pool is:

48.6% Class 1 <BR>
23.6% Class 2 <br>
27.8% Class 4 <br>


```python

# Borough composition

borough_summary = (
    master_property["borough_lien"]
    .value_counts(dropna=False)
    .rename_axis("borough")
    .reset_index(name="properties")
)

borough_summary["share"] = (
    borough_summary["properties"] /
    borough_summary["properties"].sum()
)

display(borough_summary)
```

<b>
Brooklyn and Queens Dominate the Final-Sale Pool
</B> <BR>

Brooklyn: 38.1% <BR>
Queens: 29.1% <BR>
Bronx: 13.6% <BR>
Manhattan: only 10.9% <BR>

```python
# Building class composition

building_class_summary = (
    master_property["building_class"]
    .value_counts(dropna=False)
    .head(25)
    .rename_axis("building_class")
    .reset_index(name="properties")
)

building_class_summary["share"] = (
    building_class_summary["properties"] /
    building_class_summary["properties"].sum()
)

display(building_class_summary)
```

```python
# PLUTO land use composition

landuse_summary = (
    master_property["landuse"]
    .value_counts(dropna=False)
    .rename_axis("landuse")
    .reset_index(name="properties")
)

landuse_summary["share"] = (
    landuse_summary["properties"] /
    landuse_summary["properties"].sum()
)

display(landuse_summary)
```

```python
# Residential unit distribution

master_property["unitsres"] = pd.to_numeric(
    master_property["unitsres"],
    errors="coerce"
)

master_property["unitstotal"] = pd.to_numeric(
    master_property["unitstotal"],
    errors="coerce"
)

display(
    master_property[[
        "unitsres",
        "unitstotal"
    ]].describe()
)
```

```python
# Small homes vs larger buildings

master_property["small_property_flag"] = np.where(
    master_property["unitsres"].fillna(0) <= 3,
    1,
    0
)

small_property_summary = (
    master_property["small_property_flag"]
    .value_counts(dropna=False)
    .rename_axis("small_property_flag")
    .reset_index(name="properties")
)

small_property_summary["share"] = (
    small_property_summary["properties"] /
    small_property_summary["properties"].sum()
)

display(small_property_summary)
```

B1/B2/B3 → 2-family homes <br>
A1/A2/A5 → 1-family homes <br>
R4 → condo-related <br>

## Section 2: Neighborhood vulnerability

```python
# ACS variable overview

acs_cols = [
    "median_household_income",
    "poverty_rate",
    "rent_burden_rate",
    "severe_rent_burden_rate",
    "homeownership_rate",
    "vacancy_rate",
    "age_65_plus_rate",
    "black_rate",
    "hispanic_rate",
    "asian_rate",
    "white_nonhispanic_rate",
    "bachelors_plus_rate"
]

display(master_property[acs_cols].describe())
```

```python

# Tract-level summary dataset

tract_summary = (
    master_property
    .dropna(subset=["geoid"])
    .groupby("geoid", as_index=False)
    .agg(
        final_sale_properties=("bbl", "nunique"),
        avg_open_hpd_violations=("open_hpd_violations", "mean"),
        total_open_hpd_violations=("open_hpd_violations", "sum"),
        avg_open_c_violations=("open_c_violations", "mean"),
        median_household_income=("median_household_income", "first"),
        poverty_rate=("poverty_rate", "first"),
        rent_burden_rate=("rent_burden_rate", "first"),
        severe_rent_burden_rate=("severe_rent_burden_rate", "first"),
        homeownership_rate=("homeownership_rate", "first"),
        vacancy_rate=("vacancy_rate", "first"),
        age_65_plus_rate=("age_65_plus_rate", "first"),
        black_rate=("black_rate", "first"),
        hispanic_rate=("hispanic_rate", "first"),
        asian_rate=("asian_rate", "first"),
        white_nonhispanic_rate=("white_nonhispanic_rate", "first"),
        bachelors_plus_rate=("bachelors_plus_rate", "first")
    )
)

print(tract_summary.shape)
display(tract_summary.head())
```

```python

# Neighborhoods with highest concentrations

top_tracts = (
    tract_summary
    .sort_values(
        "final_sale_properties",
        ascending=False
    )
)

display(
    top_tracts[[
        "geoid",
        "final_sale_properties",
        "poverty_rate",
        "rent_burden_rate",
        "median_household_income",
        "avg_open_hpd_violations"
    ]].head(25)
)
```

```python

# Correlation analysis

corr_cols = [
    "final_sale_properties",
    "avg_open_hpd_violations",
    "poverty_rate",
    "rent_burden_rate",
    "severe_rent_burden_rate",
    "homeownership_rate",
    "vacancy_rate",
    "age_65_plus_rate",
    "black_rate",
    "hispanic_rate",
    "white_nonhispanic_rate",
    "bachelors_plus_rate",
    "median_household_income"
]

corr_matrix = (
    tract_summary[corr_cols]
    .corr(numeric_only=True)
)

display(corr_matrix)
```

```python

# Simple vulnerability index
from scipy.stats import zscore

vulnerability_vars = [
    "poverty_rate",
    "rent_burden_rate",
    "vacancy_rate",
    "age_65_plus_rate",
    "avg_open_hpd_violations"
]

for col in vulnerability_vars:
    tract_summary[f"{col}_z"] = zscore(
        tract_summary[col].fillna(tract_summary[col].median())
    )

# Reverse-code protective variables
tract_summary["income_inverse_z"] = zscore(
    -tract_summary["median_household_income"].fillna(
        tract_summary["median_household_income"].median()
    )
)

tract_summary["homeownership_inverse_z"] = zscore(
    -tract_summary["homeownership_rate"].fillna(
        tract_summary["homeownership_rate"].median()
    )
)

tract_summary["vulnerability_index"] = (
    tract_summary[[f"{c}_z" for c in vulnerability_vars]].sum(axis=1)
    + tract_summary["income_inverse_z"]
    + tract_summary["homeownership_inverse_z"]
)

display(
    tract_summary["vulnerability_index"].describe()
)
```

```python
# Most vulnerable enforcement tracts

most_vulnerable = (
    tract_summary
    .sort_values(
        "vulnerability_index",
        ascending=False
    )
)

display(
    most_vulnerable[[
        "geoid",
        "final_sale_properties",
        "vulnerability_index",
        "poverty_rate",
        "rent_burden_rate",
        "median_household_income",
        "avg_open_hpd_violations"
    ]].head(30)
)
```

```python
# Scatterplot — vulnerability vs enforcement concentration

import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))

plt.scatter(
    tract_summary["vulnerability_index"],
    tract_summary["final_sale_properties"],
    alpha=0.6
)

plt.xlabel("Neighborhood Vulnerability Index")
plt.ylabel("Final Sale Properties")
plt.title("Enforcement Exposure and Neighborhood Vulnerability")

plt.show()
```

```python
# Borough-level vulnerability comparison

borough_vulnerability = (
    master_property
    .groupby("borough_lien", as_index=False)
    .agg(
        properties=("bbl", "nunique"),
        median_income=("median_household_income", "mean"),
        poverty_rate=("poverty_rate", "mean"),
        rent_burden_rate=("rent_burden_rate", "mean"),
        avg_hpd_violations=("open_hpd_violations", "mean")
    )
)

display(borough_vulnerability)
```
