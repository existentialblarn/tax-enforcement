### <span style="font-family:Georgia"> The Enforcement Dilemma: Keeping New York City’s Single Largest Revenue Source Stable </span>

RPT is the single largest revenue source in New York City, making up 43 percent of all City tax revenue, sustaining our schools, safety, and infrastructure.
Its predictable collections keep the City fiscally sound; therefore, City services constant. Annually, the City collects 98% of property taxes from owners citywide. 
However, every 1% of the uncollected portion -- delinquency as we call it -- is associated with over $350 million loss in annual tax revenue. With this, an effective enforcement might be necessary to ensure fairness, compliance, and the stability of providing essential City services. <br>
Every tax system has enforcement tools. For income taxes, that can mean wage garnishment or asset seizure, even jail in extreme cases. Property taxes are no different: there must be a consequence for nonpayment. <br>

Yet, administration of tax collections should also recognize that tax burdens and foreclosure risks are not distributed evenly across communities. For instance, decades of redlining, under-appraisal, and credit exclusion made homeowners in neighborhoods of color more likely to be “asset-rich but cash-poor.” 
If not acknowledged, the same enforcement tool that ensures fairness on paper may amplify inequities in practice.
<br>

Lien sale is the City's policy framework for preserving municipal fiscal stability. In place since 1996, the City purports that Lien Sale ensures tax compliance while preserving homeowner protections. It is a structured process that gives owners multiple resources — payment plans, exemptions, assistance — before penalties escalate. Fewer than 2% of liens reach foreclosure, and most are vacant land. Still, the program remains controversial in tax and housing debates; thus, it changed over time to address equity concerns.



#### <b> What Makes a Good Tax Enforcement

Nonpayment may result from the owner's decision to not pay, failure to observe due dates, or financial challenges that create hurdles to paying. Tax delinquency is also an indicator of property distress. Delinquent property tax enforcement is a powerful tool for municipalities to interrupt the cycle of decline of a property and/or to interrupt a property owner's path to deeper debt.


However, traditional approaches to property tax non-payment flags major concerns. Before the program, like other jurisdictions, the city used to sell debt to individual investors at auction or sell properties directly to private investors. <br>
Selling debt gave private investors the control over the collection process, which means they were not required to offer payment plans and no City oversight. Additionally, similar to when auctioning properties directly, investors also refuse to buy liens on the most distressed properties, making the City the long-term owner of distressed properties that are difficult and costly to maintain.

A good tax enforcement must be effective, efficient, and equitable. The goal of enforcement is to collect, not to foreclose. The measure, then, of effectiveness is yielding maximum payments and redemptions. Usually, this comes from an enforcement with teeth: motivating compliance with compelling penalties and interest, but all the while, ensuring that legacy residents, community members, and other vulnerable populations are offered accessible off-ramp programs. This includes hardship payment plans, ineligibility and non-foreclosure on small debts owed, outreach and property ownership counseling programs. <br>
More importantly, the City must control this process. But with competing City priorities, the complexity and work that comes with tax collection for more than a million properties should be well-designed. Leveraging "automatic" timeframes for each enforcement stage standardize the process within the tax collection, legal, and housing preservation's existing systems, on top of fair treatment across taxpayers, is one way to make it as cost-effective as possible.

Factoring these, the City developed a "third way" that ensures the tax collection process is​ fair and professionally administered. ​Lien sale relieves the City of taking ownership of distressed properties  foreclosed upon (prior to it the City was saddled with portfolio of around 100,000 distressed properties) while allowing property owners the chance to get back to good standing for up to 2 years and otherwise, the potential to recapture any available equity at an auction sale.

#### <b> How Lien Sale Works

![lien-process](../lien_process.png)

Lien sale begins with selecting the eligible pool: Only significant delinquencies are included; small balances and eligible owners are ineligible. Second, the Department of Finance do multi-layer outreach, mainly, a 90-, 60-, 30-, and 10-day mail notices to maximize compliance. Easy Exit, expanded payment plans, and minimum payoffs are offered throughout this process. Typically, putting 85 percent of initially eligible (those included in the 90-day mail notifications) properties back to good standing before the sale date -- that more than 25,000 properties.
<br>   
  Second, the City creates an New York City Tax Lien (NYCTL) Trust account, managed by DOF, OMB, Law, and “sells” the pool of unresolved taxes to it. The trust is authorized to sell bonds with liens pledged as a collateral, not directly selling debt to financial institutions or bondholders.
With this, the City-controlled Trust structure serves dual purposes: it provides immediate revenue to maintain the City's fiscal schedule while preserving municipal control over collection enforcement and equity safeguards. <br>
<br> The Trust hires professional servicers to collect delinquent taxes. An owner can enter a payment plan with the servicers and pay off the lien at any point, and the Trust can delay any foreclosure action. The delinquent charges collected by the Trust first pay off the issued bonds, and then once the bonds are paid off, the residual collections are remitted to the City. 
<br><br> If the debt remains unresolved, the owner enters the post-sale actions. At any time during this process, a property owner can still redeem the lien by paying off charges and fees. Property owners may also transfer foreclosed residential properties to approved nonprofits and remain as tenants at affordable rents. If charges are not paid or payment plan is not established, the Trust will initiate foreclosure procedures, typically 7 to 12 months after the sale. This can take more than 2 years to complete. Finally, if a property is sold at a tax foreclosure sale, they retain the surplus funds.





Yet, while Lien Sale attempts to address enforcement dilemmas, where fairness in rules coalesce with fairness in outcomes, there remains a quantified analysis on how effective it really is in helping vulnerable populations. Advocates maintain it continues to disproportionately affect low income, neighborhoods with high concentration of black and hispanic population, and legacy homeowners.

#### <b> Analysis Questions and Methodology

Hence, this data analysis attempts to answer the following questions:
1. What types of properties dominate the final-sale pool?
2. Do final-sale properties exhibit housing distress?
3. Are final-sale properties concentrated in neighborhoods with higher socioeconomic vulnerability?

The core question is if the City can maintain credible tax enforcement while avoiding disproportionate pressure on vulnerable housing and neighborhoods.

This analysis uses several NYC property and socioeconomic open data: Lien sale list and the Real Property Assessment Data (RPAD) from the DOF, Open housing violations from the Housing Preservation and Development (HPD), Primary Land Use and Tax Lot Output (PLUTO) from the City Planning, and Census' American Community Survey. <br><br>
It is worth noting that DOF's data does not distinguish coops and condos as individual properties; hence, their count is excluded in the analysis.

#### <b> Insights

### **1. The Final Lien Sale Pool Is Mostly Residential**

The final sale pool is heavily residential.

Class 1 properties account for approximately **49 percent** of the final sale pool, while Class 2 properties account for approximately **24 percent**. Together, residential tax classes make up nearly **three-quarters** of the properties reaching final sale status. Class 4 commercial properties account for the remaining **28 percent**.

The properties most exposed to final sale enforcement are largely connected to the city’s housing stock. The final sale stage is therefore not just a revenue collection mechanism; it is also a housing and neighborhood issue.

---

### **2. Final Sale Properties Are Concentrated in the Outer Boroughs**

Brooklyn accounts for approximately **38 percent** of final sale properties, followed by Queens at approximately **29 percent**. Together, Brooklyn and Queens represent roughly **two-thirds** of the final sale pool.

The Bronx accounts for about **14 percent**, Manhattan about **11 percent**, and Staten Island about **8 percent**.


---

### **3. The Final Sale Pool Is Small-Property Oriented**

The property characteristics reinforce the residential and outer-borough findings. Many of the most common building classes in the final sale pool are small residential or neighborhood-scale buildings, including:

- **One- and two-family homes**
- **Walk-up apartment buildings**
- **Small mixed-use residential buildings**
- **Condo-related properties**
- **Vacant or underutilized lots**

Among properties with available PLUTO unit data, the median property has only **two residential units**, and the 75th percentile has **three units**. This means that most matched final sale properties are not large apartment complexes. They are small buildings.


---

### **4. Condo-Related Properties Create a Data and Policy Complication**

The building class results also show a meaningful number of condo-related properties, including classes such as **R4**, **RG**, **RP**, **RB**, and **RS**.

This is important for two reasons.

First, condo-related records create technical matching challenges. Condo unit BBLs do not always match cleanly to PLUTO tax-lot geography, which explains part of the missing PLUTO and ACS match rate in the dataset.

Second, condo properties complicate interpretation. They are legally and administratively distinct from one-, two-, and three-family homes, but they still appear in the final sale pool as individual property records.


---

### **5. Housing Distress Is Present, but Unevenly Distributed**

The HPD violation data shows that housing distress exists within the final sale pool, but it is not evenly distributed across all properties.

Many final sale properties have no open HPD violations. However, certain multifamily and walk-up building classes show much higher average violation counts. In particular, older apartment-style building classes appear to carry elevated HPD violation burdens.

This suggests that there are at least two different types of properties in the final sale pool:

1. **Small residential properties with little visible HPD distress**
2. **A smaller subset of multifamily or walk-up buildings with significant physical distress**

This implies that some properties may simply be delinquent, while others may be part of a broader pattern of financial and physical building distress.

---

### **6. Property Values Suggest a Middle-Market Enforcement Story**

The DOF valuation data suggests that the typical final sale property is not a top high-value asset. The median full value among matched records is approximately **$509,000**, with the 75th percentile around **$770,000**.

The distribution is highly skewed. A small number of high-value properties raise the mean, but the median points to a more moderate property-value profile.


---

### **7. Neighborhood Vulnerability Overlaps with Enforcement, but Not Perfectly**

The ACS analysis shows that final sale properties are often located in neighborhoods with meaningful housing and socioeconomic pressures. Many matched properties are in tracts with substantial rent burden, moderate poverty, and varying levels of homeownership.

However, the relationship is not simple. The tracts with the most final sale properties are not always the poorest tracts. Similarly, the most vulnerable tracts do not always have the largest number of final sale properties.

Final sale concentration should not be interpreted as a direct proxy for poverty. Raw counts may reflect many factors, including:

- the number of tax lots in a tract,
- the density of small residential properties,
- the prevalence of condo records,
- local ownership patterns,
- building age,
- and socioeconomic vulnerability.

---

### **8. Boroughs Show Different Versions of the Enforcement Dilemma**

The borough-level results suggest that the enforcement dilemma looks different across the city.

**Brooklyn** has the largest number of final sale properties. This makes it the center of the enforcement volume story.

**Queens** also has a large final sale footprint, but its measured poverty and HPD distress levels are lower than the Bronx and Brooklyn in this dataset. This suggests a more mixed enforcement profile.

**The Bronx** has fewer final sale properties than Brooklyn or Queens, but it shows stronger overlap with poverty, rent burden, and HPD violations. This makes the Bronx especially important for understanding the intersection of enforcement and vulnerability.

**Manhattan** accounts for a smaller share of the final sale pool.

**Staten Island** has the smallest count but still represents part of the small-property enforcement landscape.

**Key takeaway:**  
Brooklyn appears to be the highest-volume final sale borough, while the Bronx appears to show the strongest overlap between enforcement, housing distress, and socioeconomic vulnerability.

---

### **9. The Core Enforcement Dilemma**

The central dilemma is not whether the City should collect unpaid property taxes. Property tax enforcement is essential to maintaining the stability of the City’s largest revenue source.

The dilemma is that the most serious stages of enforcement do not fall largely on large commercial or institutional properties. They  reach small residential properties, outer-borough neighborhoods, and, in some cases, physically distressed buildings and socioeconomically vulnerable areas.

This creates a difficult policy balance.

On one hand, the City needs credible enforcement to protect revenue stability, maintain fairness for compliant taxpayers, and prevent delinquency from becoming normalized.

On the other hand, aggressive enforcement can intersect with housing instability, deferred maintenance, small-property ownership challenges, and neighborhood vulnerability.

---

### **Overall Conclusion**

This analysis finds that New York City’s final lien sale pool is heavily residential, outer-borough, and small-property oriented. While final sale concentration is not simply a proxy for poverty, it often overlaps with broader indicators of housing and neighborhood vulnerability, especially in places where small residential properties, HPD violations, and socioeconomic stress coincide.

The policy challenge is therefore not simply whether the City should enforce property tax collection. The harder question is how the City can preserve the credibility and stability of its largest revenue source while recognizing that final-stage enforcement can affect small properties, distressed housing stock, and vulnerable neighborhoods.

**In short:**  
The City cannot afford not to enforce its property tax system, but enforcement is not socially or spatially neutral.
